```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import numpy as np # linear algebra
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
import matplotlib.pyplot as plt

import csv
df=pd.read_csv("/kaggle/input/voicecsv/voice.csv")
print(df)
df.shape
df.isnull().sum()



y=df.label
y1=y

print(y)
x=df.drop('label',axis=1)
print(x)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
m1=LogisticRegression(max_iter=1000)
m1.fit(x_train,y_train)
print('Training score',m1.score(x_train,y_train))
print('Test score',m1.score(x_test,y_test))
m1.fit(x_test,y_test)
ypred=m1.predict(x_test)

cm=metrics.confusion_matrix(y_test,ypred)
print("Confusion matrix for model 1:",cm)
print(metrics.classification_report(y_test,ypred))
m2=DecisionTreeClassifier(criterion='gini',max_depth=4,min_samples_split=12)
m2.fit(x_train,y_train)
print('Training score',m2.score(x_train,y_train))
print('Test score',m2.score(x_test,y_test))
ypred2=m2.predict(x_test)
cm=metrics.confusion_matrix(y_test,ypred2)
print(metrics.classification_report(y_test,ypred2))
print("Confusion matrix for model 2",cm)
print('Training score',m2.score(x_train,y_train))
print('Test score',m2.score(x_test,y_test))
m3=RandomForestClassifier(n_estimators=100,criterion='entropy',random_state=0)
m3.fit(x_train,y_train)
ypred3=m3.predict(x_test)
print(ypred3)
cm=metrics.confusion_matrix(y_test,ypred3)
print("Confusion matrix for model 3",cm)
print(metrics.classification_report(y_test,ypred3))
print('Training score',m3.score(x_train,y_train))
print('Test score',m3.score(x_test,y_test))

m4=SVC(kernel='linear',C=1)
m4.fit(x_train,y_train)
ypred4=m4.predict(x_test)
print(ypred4)
cm=metrics.confusion_matrix(y_test,ypred4)
print("Confusion matrix for model 4",cm)
print(metrics.classification_report(y_test,ypred4))
print('Training score',m4.score(x_train,y_train))
print('Test score',m4.score(x_test,y_test))
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
knn.predict(x_test)
print(knn.predict(x_test))
cm=metrics.confusion_matrix(y_test,knn.predict(x_test))
print(metrics.classification_report(y_test,knn.predict(x_test)))
print('Training score',knn.score(x_train,y_train))
print('Test score',knn.score(x_test,y_test))
print("Since RandomForestClassifer has the best test score, it is the most accurate model")
a=(list(y)).count('male')
b=(list(y)).count('female')
print(a,b)
genders=['male','female']
data=[a,b]
plt.pie(data,labels=genders,autopct='%1.0f%%')
print("accuracy score for LogisticRegression",accuracy_score(y_test, m1.predict(x_test)))
print("accuracy score for DecisionTree",accuracy_score(y_test, m2.predict(x_test)))
print("accuracy score for RandomForestClassifier",accuracy_score(y_test, m3.predict(x_test)))
print("accuracy score for KNN",accuracy_score(y_test, knn.predict(x_test)))
print("accuracy score for SVC",accuracy_score(y_test, m4.predict(x_test)))
print("Since RandomForestClassifer has the best accuracy_score, it is the most accurate model")
```
